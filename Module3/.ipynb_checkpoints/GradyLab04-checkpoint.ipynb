Data Analysis

```python
con = [0.00, 0.00001, 0.00002, 0.00004, 0.00006, 0.00008, 0.0001]
absorb = [0.0, 0.137, 0.160, 0.324, 0.471, 0.5, 0.428]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.polyfit(con, absorb, 1)
p = np.poly1d(z)

correlation_matrix = np.corrcoef(con, absorb)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2

print(r_squared)
```

```python
plt.figure(figsize=(16,10))
plt.scatter(con,absorb,label='Standard Solutions')
plt.plot(con, p(con),'r',label="Trendline y=%.2fx+%.3f"%(z[0],z[1]))
plt.xlabel('Concentration (g/mL)',fontsize=20)
plt.ylabel('Absorbancy (AU)',fontsize=20)
plt.legend(fontsize=16)
plt.annotate("R-squared=0.826",(0.0000,0.45),fontsize=20)
plt.title('Absorbancy of a Solution As a Function of Concentration',fontsize=25)
plt.savefig('Lab04Graph.pdf')
```
