### 1. Introduction

In the accompanying notebook [mod02-CarbonModel1.ipynb](./mod02-CarbonModel1.ipynb) and in class we developed a relatively simple model of global carbon balance that assumed that tracked carbon in the atmosphere and biosphere. The system of equations we developed to describe this system are shown below,

$$
\begin{align*}
\frac{dM_1}{dt} &= F_{21} - F_{12}\\
&= k_{21}M_2 - k_{12}M_1
\end{align*}
$$

and,

$$
\begin{align*}
\frac{dM_2}{dt} &= F_{12} - F_{21}\\
&= k_{12}M_1 - k_{21}M_2
\end{align*}
$$

where $M_1$ is the mass of carbon in the atmosphere (GtC), $M_2$ is the mass of carbon in the biosphere (GtC), $F_{12}$ is the flux of carbon from the atmosphere to the biosphere due to photosynthesis (GtC/yr), and $F_{21}$ is the flux of carbon from the biosphere to the atmosphere due to decay and respiration (GtC/yr).

In that notebook we develop a numerical model to describe that system, run it for given initial conditions and with no external inputs of carbon (we call this an "unforced" scenario) and compare it to the analytical solution. We observed that as the time step increased, we departed further from the analytical solution.

Then you will examine the impact of anthropogenic emissions from 1751-2008 on the biosphere and atmosphere, and how alternative future scenarios of emissions mitigation affect your system.

__Problem A: Modify and rerun the model__

1. Modify the model so that the photosynthesis flux includes the mass of carbon in the land (i.e., $M_2$). Use the following parameters:

  * $M_1^0$ = 1100.0 
  * $M_2^0$ = 300.0 
  * $k_{12}$ = 0.0003 
  * $k_{21}$ = 0.1  

2. Repeat the "unforced" (i.e., no anthropogenic emissions) simulation above, but run the model for `t = 100` years with a daily time step (i.e., `dt = 1/365`)
3. Use the `print()` command to print out the value of $M_1$ and $M_2$ at the end of the simulation



```python
import numpy as np
import matplotlib.pyplot as plt

#M1i = initial atmosphere carbon mass
M1i = 1100.0
#M2i = initial biosphere carbon mass
M2i = 300.0

#k12 = amount of carbon flux (residence time) from the atmosphere to biosphere - photosynthesis
k12 = 0.0003 
#k21 = amount of carbon flux (residence times) from the biosphere to the atmosphere
k21 = 0.1

ti = 0.0 #initial time
tf = 100.0 #final time step
dt = 1/365 #step size

```

```python
t = np.arange(ti,tf+dt,dt) #time intervals

Nt = t.size #number of time elements on x-axis

print('t has '+str(Nt)+' time steps')
```

```python
M1 = np.zeros((Nt,)) #returns the array filled with zeros
M2 = np.zeros((Nt,))
```

```python
for i in np.arange(Nt): #for loop to apply 'k' steps to the initial carbon concentation over 21 time steps
    if (i==0):
# Sets initial concentrations, from the np.zeros function
        M1[i] = M1i
        M2[i] = M2i
#applies the mass balance function over all time steps 
    else:
        dM1dt = k21*M2[i-1] - k12*M1[i-1]*M2[i-1]
        dM2dt = k12*M1[i-1]*M2[i-1] - k21*M2[i-1]
#Integrates the function    
        M1[i] = M1[i-1] + dM1dt*dt
        M2[i] = M2[i-1] + dM2dt*dt
```

```python
plt.figure(figsize=(8.5,11))
plt.rcParams.update({'font.size': 16})
plt.title('Box Model Carbon Model')
#Plot for carbon amount in the atmosphere
plt.subplot(2,1,1)
plt.plot(t,M1,'bo', label='Carbon flux (GT/year)')
#plt.plot(t,M1_anlt,'k-', label='Analytical Carbon Flux (GT/year)')
plt.xlabel('Time (years)')
plt.ylabel('Mass of Carbon in Atmosphere (GT)')
plt.legend()
#Plot for carbon amount in the biosphere
plt.subplot(2,1,2)
plt.plot(t,M2,'bo', label='Carbon Flux GT/year')
#plt.plot(t,M2_anlt,'k-', label='what is this #2b?')
plt.xlabel('Time (years)')
plt.ylabel('Mass of Carbon in Biosphere (GT)')
plt.legend()

#Initial conditions for historical data
print('M1 =',M1[-1])

print('M2 =',M2[-1])
```

__Problem C: Examine the effects of mitigation__

1. Load the alternative emission scenarios contained in the file `.csv`, which represent three pathways to net zero emissions: (a) net zero emissions by the year 2030 (column 2), (c) net zero emissions by 2050 (column 3), and zero emissions growth (i.e., constant emissions) starting in 2008 (column 4).
2. Starting from the initial conditions ($M_1^0$ and $M_2^0$) corresponding to the final conditions noted in part __B(4)__ above, run the model three times, one for each scenario. Create the following plots:

  * Mass of C in the land versus time (all three scenarios on one plot)
  * The change in mass of C in the atmosphere vs time

```python
future_emissions_file = 'EmissionsMitigationScenarios.2008_2100.csv'
#Load future projection datat
data = np.loadtxt(future_emissions_file, delimiter=',', skiprows=1)
year = data[:,0] #sorts data into 3 projections with time (year)
cflux_nz2030 = data[:,1]
cflux_nz2050 = data[:,2]
cflux_zeg = data[:,3]

data
```

```python
#M1i = initial atmosphere carbon mass
M1i = M1[-1]
#M2i = initial biosphere carbon mass
M2i = M2[-1]

#k12 = amount of carbon flux (residence time) from the atmosphere to biosphere - photosynthesis
k12 = 0.0003
#k21 = amount of carbon flux (residence times) from the biosphere to the atmosphere
k21 = 0.1


ti = year[0] #initial time
tf = year[-1] #final time ste
dt = 1/365 #step size

t = np.arange(ti,tf+dt,dt)

cflux_nz2030_t = np.interp(t, year, cflux_nz2030) #interpolates projections from yearly to daily. 
cflux_nz2050_t = np.interp(t, year, cflux_nz2050)
cflux_zeg_t = np.interp(t, year, cflux_zeg)

```

```python


Nt = t.size #number of time elements on x-axis

print('t has '+str(Nt)+' time steps')
```

```python
M1_nz2030 = np.zeros((Nt,)) #returns the array filled with zeros
M2_nz2030 = np.zeros((Nt,))

M1_nz2050 = np.zeros((Nt,)) #returns the array filled with zeros
M2_nz2050 = np.zeros((Nt,))

M1_zeg = np.zeros((Nt,)) #returns the array filled with zeros
M2_zeg = np.zeros((Nt,))
```

```python
for i in np.arange(Nt): #for loop to apply 'k' steps to the initial carbon concentation over 21 time steps
    if (i==0):
# Sets initial concentrations, from the np.zeros functions for all 3 projections
        M1_nz2030[i] = M1i
        M2_nz2030[i] = M2i
        
        M1_nz2050[i] = M1i
        M2_nz2050[i] = M2i
        
        M1_zeg[i] = M1i
        M2_zeg[i] = M2i
#applies the mass balance function over time steps for all projections 
    else:
        M1_nz2030dM1dt = k21*M2_nz2030[i-1] - k12*M1_nz2030[i-1]*M2_nz2030[i-1]+ cflux_nz2030_t[i-1]
        M2_nz2030dM2dt = k12*M1_nz2030[i-1]*M2_nz2030[i-1] - k21*M2_nz2030[i-1]
        
        M1_nz2050dM1dt = k21*M2_nz2050[i-1] - k12*M1_nz2050[i-1]*M2_nz2050[i-1]+ cflux_nz2050_t[i-1]
        M2_nz2050dM2dt = k12*M1_nz2050[i-1]*M2_nz2050[i-1] - k21*M2_nz2050[i-1]
        
        M1_zegdM1dt = k21*M2_zeg[i-1] - k12*M1_zeg[i-1]*M2_zeg[i-1]+ cflux_zeg_t[i-1]
        M2_zegdM2dt = k12*M1_zeg[i-1]*M2_zeg[i-1] - k21*M2_zeg[i-1]
#creates the integrated function  from projections  
        M1_nz2030[i] = M1_nz2030[i-1] + M1_nz2030dM1dt*dt
        M2_nz2030[i] = M2_nz2030[i-1] + M2_nz2030dM2dt*dt
        
        M1_nz2050[i] = M1_nz2050[i-1] + M1_nz2050dM1dt*dt
        M2_nz2050[i] = M2_nz2050[i-1] + M2_nz2050dM2dt*dt
        
        M1_zeg[i] = M1_zeg[i-1] + M1_zegdM1dt*dt
        M2_zeg[i] = M2_zeg[i-1] + M2_zegdM2dt*dt
```

```python
plt.figure(figsize=(8.5,11))
plt.rcParams.update({'font.size': 16})
plt.title('What is this?')
#Plots atmospheric carbon with each respective projection.
plt.subplot(2,1,1)
plt.plot(t,M1_nz2030, label='NZ 2030 (GT/year)')
plt.plot(t,M1_nz2050, label='NZ 2050 (GT/year)')
plt.plot(t,M1_zeg, label='ZEG (GT/year)')
plt.xlabel('Time (years)')
plt.ylabel('Mass of Carbon in Atmosphere (GT)')
plt.legend()

#Plots biospheric carbon with each respective projection.
plt.subplot(2,1,2)
plt.plot(t,M2_nz2030, label='NZ 2030 (GT/year)')
plt.plot(t,M2_nz2050, label='NZ 2050 (GT/year)')
plt.plot(t,M2_zeg, label='ZEG (GT/year)')
plt.xlabel('Time (years)')
plt.ylabel('Mass of Carbon in Biosphere (GT)')
plt.legend()
plt.show()
```
